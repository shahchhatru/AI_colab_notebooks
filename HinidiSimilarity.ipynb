```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("l3cube-pune/hindi-sentence-similarity-sbert")
```

```python
hindi_sentences = [
    "आज मौसम बहुत सुहाना है।",
    "कल मौसम थोड़ा ठंडा था। ",  # Similar meaning to previous sentence
    "मुझे किताबें पढ़ना पसंद है। ",
    "वह क्रिकेट खेलना पसंद करती है।",  # Dissimilar meaning to previous sentences
]

nepali_sentences = [
    "आजको मौसम निकै राम्रो छ। ",
    "हिजोको मौसम अलि चिसो थियो।",  # Similar meaning
    "मलाई किताबहरू पढ्न मन पर्छ। ",
    "उनी गीत सुन्न मन पराउँछिन्।",  # Dissimilar meaning
]
```

```python
embeddings_hindi = model.encode(hindi_sentences)
embeddings_nepali=model.encode(nepali_sentences)
```

```python
from sentence_transformers import util

import torch

def recommendation(input_paper,rec_model,embeddings,sentences):
    # Calculate cosine similarity scores between the embeddings of input_paper and all papers in the dataset.
    cosine_scores = util.cos_sim(embeddings, rec_model.encode(input_paper))

    # Get the indices of the top-k most similar papers based on cosine similarity.
    top_similar_papers = torch.topk(cosine_scores, dim=0, k=4, sorted=True)

    # Retrieve the titles of the top similar papers.
    papers_list = []
    for i in top_similar_papers.indices:
        papers_list.append(sentences[i.item()])

    return papers_list
```

```python
input_text ="आज मौसम खिली खिली है"
recommend_papers = recommendation(input_text,model,embeddings_hindi,hindi_sentences)


print("We recommend to read this text............")
print("=============================================")
for paper in recommend_papers:
    print(paper)
```

```python
input_text ="आज हिमाल टल्कै छन्, आकाश भने सफा छ "
recommend_papers = recommendation(input_text,model,embeddings_nepali,nepali_sentences)


print("We recommend to read this text............")
print("=============================================")
for paper in recommend_papers:
    print(paper)
```
